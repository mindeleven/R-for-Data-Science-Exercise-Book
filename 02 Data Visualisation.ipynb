How to visualise data with the ggplot2 package.

For a first look at the ggplot2 package we want to answer the question if cars with big engines use more fuel than cars with small engines. To answer this question ggplot2's mpg data frame is used.

```r
mpg
```

The variables of interest here are displ (engine size in litres) and hwy (fuel efficiency in miles per gallon). A low fuel efficiency means that a car consumes more fuel.

### Creating a simple ggplot for exploration
The ggplot function creates a coordinate system and the geom_point function adds a layer of points that create a scatterplot.
ggplot() takes the dataset as an argument and the geom() function takes a mapping argument (paired with 'aes') that defines the variables.
The resulting graph shows a negative relationship, i.e. cars with big engines have a lesser fuel efficiency.

```r
ggplot(data = mpg) +
  geom_point(mapping = aes(x = displ, y = hwy))
```
